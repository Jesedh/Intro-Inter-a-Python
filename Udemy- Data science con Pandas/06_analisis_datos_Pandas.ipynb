Ejercicio: a partir del dataframe con la informacion de trafico peatonal (5.5), verificar el numero de registros y a continuacion eliminar duplicados definiendo duplicado como aquellos registros que tienen los valores de todas las columnas iguales:
* ¿Cuantos valores duplicados existian?
* Eliminar los valores erroneos (NaN) del dataframe
* ¿Cuantos valores NaN existian?
* ¿Que porcentaje de los datos originales eran erroneos?

```python
lec_peatones = pd.read_csv(r"./Datasets/5.5/PEATONES_2020_mod.csv", delimiter=";", header=0, encoding="ISO-8859-1")
peatones = pd.DataFrame(lec_peatones)

num_peatones = peatones.shape[0]
print("Numero de datos iniciales:", num_peatones)
# ¿Cuantos valores duplicados existian?
peatones.drop_duplicates(inplace=True)
num_peatones_new = peatones.shape[0]
val_dup = num_peatones - num_peatones_new
print(f"Habia {val_dup} valores duplicados en el DataFrame")
# Eliminar los valores erroneos (NaN) del dataframe, ¿Cuantos valores NaN existian?
peatones.dropna(inplace=True)
num_peatones_err = peatones.shape[0]
val_erroneos = num_peatones_new - num_peatones_err
print(f"Habia {val_erroneos} valores erroneos en el data frame")
# ¿Que porcentaje de los datos originales eran erroneos?
porcen_datos_err = (val_erroneos * 100) / num_peatones
print("El porcentaje de datos errones de acuerdo al numero de datos originales es: " + str(porcen_datos_err))
```

Interpolacion de datos: Obtener un nuevo valor a partir del conjunto de datos

Nos sirve cuando tenemos valores erroneos por ejemplo:
1. Un valor negativo en la clase de num_alumnos 
2. Valores NaN dentro de un registro pero no queremos borrarlo ya que consideramos las demas columnas que no tienen NaN podrian darnos informacion importante en nuestro analisis de datos
3. Si necesitamos mayor granularidad en nuestros datos

¿Que estrategias tenemos? : fillna y interpolate

```python
datos = np.array([[12.0,3.0,20.0,14.0],
        [4.0,2.0,16.0,3.0],
        [5.0,54.0,np.nan,np.nan],
        [np.nan,3.0,3.0,np.nan],
        [1.0,np.nan,8.0,6.0]])

indices = np.array([0,1,2,3,4])
columnas = np.array(["A","B","C","D"])
valores = pd.DataFrame(datos, columns= columnas, index = indices)

# metodo .filna() para la interpolacion de datos
valores["A"].fillna(valores["A"].median(), inplace=True)

# metodo interpolate()
valores = valores.interpolate(method="linear", limit_direction="forward")
valores

# Si queremos insertar valores nan usamos : df[df<0]=np.nan
```

```python
iris = pd.read_csv(r"./datasets/5.12/iris_dataset.csv", encoding="ISO-8859-1", delimiter=",", header=0)

pi = np.pi
iris["a_lon_sep"] = iris["longitud_sépalo"].apply(lambda x:x/2)
iris["b_anc_sep"] = iris[" anchura_sépalo"].apply(lambda x:x/2)
iris["a_lon_pet"] = iris["longitud_pétalo"].apply(lambda x:x/2)
iris["b_anc_pet"] = iris["anchura_pétalo"].apply(lambda x:x/2)
# uan ves calculados los semiejes del elipse a considerar procedemos a calcular el area del petalo y sepalo
# iris["area"] = iris["a_lon_sep"]*iris["b_anc_sep"]
iris["area_sep"] = iris.apply(lambda x:x["a_lon_sep"] * x["b_anc_sep"] * pi, axis=1)
iris["area_pet"] = iris.apply(lambda x:x["a_lon_pet"] * x["b_anc_pet"]* pi, axis=1)
iris
```

```python
iris["area_pet"] = iris.apply(lambda x:x["longitud_pétalo"]*x["anchura_pétalo"]*pi, axis=1)

# ¿Cual es la mediana del area del petalo?
mediana_pet = iris["area_pet"].median()
print(f"La mediana del area del petalo es: {mediana_pet}")
# ¿Que valor del area solamente es superado por el 10% de los petalos?
percentil_10 = iris["area_pet"].quantile(.9)
print(f"El percentil 10% de la columna area_petalo: {percentil_10}")
mayores_10 = pd.DataFrame(iris["area_pet"][iris["area_pet"] > percentil_10])
mayores_10
```

```python
# Exite correlacion entre la longitud y la anchura de los petalos?
iris.corr(method="pearson")
```

```python
correlacion = np.corrcoef(iris["longitud_pétalo"], iris["anchura_pétalo"])
print(f"Existe una correlacion muy fuerte entre la longitud y la anchura de los petalos, y es de : {correlacion[0][1]}")
```

Combinacion de dataframes(merge)

**Modos de juntar las tablas**

sean A y B tablas
* INNER JOIN : A Interseccion B : interseccion entre tablas
* OUTER JOIN : A U B : union de tablas
* LEFT OUTER JOIN : toda la tabla izquierda pero con la interseccion de la tabla de la derecha
* RIGHT OUTER JOIN : toda la tabla derecha pero con la interseccion de la tabla de la izquierda

```python
info_pais = {"pais":["Brasil", "Mexico", "España", "Peru", "Colombia", "Venezuela"],
            "capital":["Brasilia", "Ciudad de Mexico", "Madrid", "Lima", "Bogota", "Caracas"],
            "superficie":[8.5, 1.9 ,1.2 ,0.5 ,1.1 ,0.9]}

df_pais_info = pd.DataFrame(info_pais)
df_pais_info
```

```python
df_peatones = pd.read_csv(r"./Datasets/5.18/PEATONES_2020_mod.csv", delimiter=";", encoding="ISO-8859-1")
df_peatones
```

```python
df_peatones.DISTRITO.unique()
```

```python
df_pob = pd.read_csv(r"./Datasets/5.18/Distrito_poblacion.csv", delimiter=";", encoding="ISO-8859-1")
df_pob
```

```python
# Realizar una combinacion de tipo inner de ambos df creamndo df_combinado utilizando .merge()
df_combinacion = df_peatones.merge(df_pob, on="DISTRITO", how="inner")
df_combinacion
```

```python
# KPI : indicadores de calidad o indicadores clave de negocio
df_combinacion["kpi_peat_pob"] = df_combinacion["PEATONES"] / df_combinacion["Población_Distrito"]
df_combinacion
```

```python
# 2. Realizar un groupby para verificar el valor maximo de KPI en cada distrito
df_combinacion.groupby("DISTRITO")["kpi_peat_pob"].agg("max")
```
