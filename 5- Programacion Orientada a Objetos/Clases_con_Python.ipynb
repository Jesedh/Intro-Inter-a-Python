```python
class Personaje: # creamos la clase y le asiganamos un nombre
    # los atributos y los metodos se definen como funciones
    def __init__(self, nombre, fuerza, inteligencia, defensa, vida): # definimos el metodo constructor que tendra los atributos de la clase
        self.nombre = nombre # self: hace referencia a si mismo
        self.fuerza = fuerza
        self.inteligencia = inteligencia
        self.defensa = defensa
        self.vida = vida
    
    def atributos(self): # creamos un metodo de la clase para desplegar los atributos
        print(self.nombre, ":", sep="")
        print("·Fuerza:", self.fuerza)
        print("·Inteligencia:", self.inteligencia)
        print("·Defensa:", self.defensa)
        print("·Vida:", self.vida)

    def subir_nivel(self, fuerza, inteligencia, defensa): # metodo de la clase
        self.fuerza = self.fuerza + fuerza
        self.inteligencia = self.inteligencia + inteligencia
        self.defensa = self.defensa + defensa

    def esta_vivo(self):
        return self.vida > 0

    def morir(self):
        self.vida = 0
        print(self.nombre, "ha muerto")

    def daño(self, enemigo):
        return self.fuerza - enemigo.defensa

    def atacar(self, enemigo):
        daño = self.daño(enemigo)
        enemigo.vida = enemigo.vida - daño
        print(self.nombre, "ha realizado", daño, "puntos de daño a", enemigo.nombre)
        if enemigo.esta_vivo():
            print("Vida de", enemigo.nombre, "es", enemigo.vida)
        else:
            enemigo.morir()

class Guerrero(Personaje): # heredamos los atributos y clases de la clase padre (Personaje) asi a guerrero le podemos asignar aun mas atributos y clases
    
    def __init__(self, nombre, fuerza, inteligencia, defensa, vida, espada): # definimos el metodo constructor de los atributos de la clase guerrero
        super().__init__(nombre, fuerza, inteligencia, defensa, vida) # Incluimos los atributos de la clase 
        self.espada = espada

    def cambiar_arma(self):
        opcion = int(input("Elige un arma: (1) Acero Valyrio, daño 8. (2) Matadragones, daño 10"))
        if opcion == 1:
            self.espada = 8
        elif opcion == 2:
            self.espada = 10
        else:
            print("Número de arma incorrecta")

    def atributos(self):
        super().atributos()
        print("·Espada:", self.espada)

    def daño(self, enemigo):
        return self.fuerza*self.espada - enemigo.defensa

class Mago(Personaje):

    def __init__(self, nombre, fuerza, inteligencia, defensa, vida, libro):
        super().__init__(nombre, fuerza, inteligencia, defensa, vida)
        self.libro = libro

    def atributos(self):
        super().atributos()
        print("·Libro:", self.libro)

    def daño(self, enemigo):
        return self.inteligencia*self.libro - enemigo.defensa


def combate(jugador_1, jugador_2): # simulamos un combate entre personajes con una funcion
    turno = 0
    while jugador_1.esta_vivo() and jugador_2.esta_vivo():
        print("\nTurno", turno)
        print(">>> Acción de ", jugador_1.nombre,":", sep="")
        jugador_1.atacar(jugador_2)
        print(">>> Acción de ", jugador_2.nombre,":", sep="")
        jugador_2.atacar(jugador_1)
        turno = turno + 1
    if jugador_1.esta_vivo():
        print("\nHa ganado", jugador_1.nombre)
    elif jugador_2.esta_vivo():
        print("\nHa ganado", jugador_2.nombre)
    else:
        print("\nEmpate")

personaje_1 = Guerrero("Guts", 20, 10, 4, 100, 4) # creamos una objeto de la clase guerrero y le asignamos atributos
personaje_2 = Mago("Vanessa", 5, 15, 4, 100, 3) # creamos una objeto de la mago y le asignamos atributos

personaje_1.atributos() # mandamos a llamar a el metodo atributos del objeto personaje_1
personaje_2.atributos() # mandamos a llamar a el metodo atributos del objeto personaje_2

combate(personaje_1, personaje_2) # mandamos a llamar a la funcion combate y observamos el desempeño de los personaes creados.
```
