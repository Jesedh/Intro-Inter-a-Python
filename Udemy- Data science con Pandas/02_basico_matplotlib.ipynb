Use Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
# creamos dos arreglos que concatenaremos
uno = np.array([2015, 2016, 2017])
dos =np.array ([2018, 2019, 2020])

año = np.concatenate((uno, dos))
print(año)

población = [2.8, 2.9, 3, 3.2, 3.5, 4]

# linear plot - grafico de linea
# siendo eje x=año, y=poblacion
plt.plot(año, población)
plt.show()
```

```python
# Scatter plot - diagrama de dispersion
plt.scatter(año,población)
plt.show()
```
